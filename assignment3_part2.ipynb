## Task 4 - Adding LIX to the Text

A way to calculate how easy or difficult a text is in Swedish is to calculate a _readability index_ (or in Swedish, _läsbarhetsindex_ (LIX)) for the text. Copy and paste the code you did previously for Gösta Berlings Saga, but now you should extend the functionality to also calculate a LiX score for it. 

The Swedish readability index (LiX) is calculated using the following formula:


$\text{LiX} = \frac{\text{total number of words}}{\text{number of sentences}} + \left(\frac{\text{number of long words}}{\text{total number of words}} \times 100 \right)$


Where:
- A **long word** is defined as a word with more than 6 letters.
- A **sentence** is typically counted by punctuation marks like `.`, `!`, or `?`.

The score given can be interpreted as the table below shows:

| **LiX Score** | **Readability Level**           | **Description**                                            |
|---------------|----------------------------------|------------------------------------------------------------|
| 0 - 24        | **Very Easy**                    | Suitable for children's books or simple texts.              |
| 25 - 34       | **Easy**                         | Suitable for a general audience with basic reading skills.  |
| 35 - 44       | **Medium**                       | Typical readability for novels or newspapers.               |
| 45 - 54       | **Difficult**                    | Suitable for academic texts or technical documents.         |
| 55+           | **Very Difficult**               | Suitable for specialized literature or scientific papers.   |
