
````python
import random
lista = []
# Creating list with 100 random integers between 1 and 100.
for i in range(100):
    tal = random.randint(1,100)
    lista.append(tal)
# We make the list into a set, this will remove all duplicate integers.
unik = set(lista)
counter = 0
lista = list(unik)
print(f'Number of unique integers: {len(lista)}')
print('Different integers:')
# counter determines if the next integer is going to be printed on a new line or not
# so that there can be 10 integers on each line.
for i in lista:
    counter += 1
    # We move the number to the right by 2 width numbers when it is only one digit.
    if i < 10:
        print("{:>2}".format(i), end=' ')
    elif counter % 10 != 0 and counter != 0:
        print(i,end= ' ')
    elif counter % 10 != 0 and i <= 9:
        "{:>15}".format(i)
    else:
        print(i)   
````

## Task 3 - Working with Words

Included in the assignment is the file `GostaBerlingsSaga.txt` which is an old book by the Swedish author Selma Lagerlöf. The book is now out of copyright and free to use, which is what you will do in this assignment. Start from the book and the case study called Occurrences of Words (14.7, page 487 in the latest edition) and create a program that counts the number of words in the text and checks the top ten most common words. See below for an example run of the program. Notice that you do not need to get _exactly_ the same numbers, plus or minus about 50 words is fine.

```

på	1767
```

````python
# Python program to read 
# file word by word
from collections import Counter
dic = {}
lista = []
# opening the text file
with open('GostaBerlingsSaga.txt','r') as file:  
    # line is the entire txt file in original form
    for line in file:   
        # Each word on a new line
        for word in line.split():
            # Contain only alphanumeric characters
            word = ''.join(letter for letter in word if letter.isalnum())
            # Lowercase
            word = word.lower()
            lista.append(word)
print('Information about Gösta Berlings saga')
# lista contains all words so the length of lista will be number of words.
print(f'Total number of words: {len(lista)}')
print('Ten most common words:')
# Creating a dictionary where the key is a word and value is how many times it appears.
for letter in lista:
    dic[letter] = dic.get(letter, 0) + 1
# We use Counter to count the most common words, however this prints out the entire dictionary.
counter = Counter(dic)
# This for loop only prints the first 10 words and doesn't do it inside the dictionary.
# k is the key and v is the value.
for k, v in counter.most_common(10):
    print(f'{k}:  {v}')
````

The Swedish readability index (LiX) is calculated using the following formula:


$\text{LiX} = \frac{\text{total number of words}}{\text{number of sentences}} + \left(\frac{\text{number of long words}}{\text{total number of words}} \times 100 \right)$


Where:
- A **long word** is defined as a word with more than 6 letters.
- A **sentence** is typically counted by punctuation marks like `.`, `!`, or `?`.

The score given can be interpreted as the table below shows:

| **LiX Score** | **Readability Level**           | **Description**                                            |
|---------------|----------------------------------|------------------------------------------------------------|
| 0 - 24        | **Very Easy**                    | Suitable for children's books or simple texts.              |
| 25 - 34       | **Easy**                         | Suitable for a general audience with basic reading skills.  |
| 35 - 44       | **Medium**                       | Typical readability for novels or newspapers.               |
| 45 - 54       | **Difficult**                    | Suitable for academic texts or technical documents.         |
| 55+           | **Very Difficult**               | Suitable for specialized literature or scientific papers.   |
